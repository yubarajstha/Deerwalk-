```python
import matplotlib.pyplot as plt

data = [75,89,100,65,80,85,90,70,78]
plt.boxplot(data)
plt.title('Test Score Distribution')
plt.xlabel("Score")
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Apple', 'Bananas', 'Cherries', 'Dates']
sizes = [25, 30, 20, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=0)
plt.title('Fruit Distribution')
plt.show()
```

```python
catesgories = ['Math','Science','English','History']
scores = [88,90,80,75]
plt.bar(catesgories , scores, color = 'skyblue')
plt.title('student scores')
plt.xlabel('subjects')
plt.ylabel('scores')
plt.show()
```

```python
x = [ 1,2,3,4,5]
y= [5,4,6,7,5]
plt.scatter(x,y, color = 'red', marker = 'o')
plt.title('Simple Scatter Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True)
plt.show()
```
