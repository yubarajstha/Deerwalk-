```python
import numpy as np
import pandas as pd
filename="Canada.csv"
df_can= pd.read_csv(filename)
df_can.head()
```

```python
df_can.set_index('Country', inplace= True)
```

```python
df_can.head(5)
```

```python
years = list(map(str,range(1980, 2014)))
years
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.__version__
```

```python
print(plt.style.available)
mpl.style.use(['ggplot'])  # optionla : for ggplot_like style
```

```python
haiti= df_can.loc['Haiti',years]
haiti.head(5)
```

```python
haiti.plot()
```

```python
haiti.plot(kind='line')
plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.text(2000, 6000, '2010 Earthquake')
plt.show()
```

```python
df_CI= df_can.loc[['India', 'China'], years]
df_CI
```

```python
df_CI.plot()
```

```python
df_CI = df_CI.transpose()
df_CI.head()
```

```python
df_CI.index = df_CI.index.map(int)
df_CI.plot(kind = 'line')
plt.title('Immigranys from china and india')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
df_al= df_can.loc[['Algeria', 'Albania'], years]
df_al
```

```python
df_al.plot()
```

```python
df_al = df_al.transpose()
df_al.head()
```

```python
df_al.index = df_al.index.map(int)
df_al.plot(kind = 'line')
plt.title('Immigranys from Algeria and Albania')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
df_can.sort_values(by= "Total", ascending = False , axis = 0 , inplace = True)
df_top5= df_can.head(5)
df_top5 = df_top5[years].transpose()
print(df_top5)

df_top5.index = df_top5.index.map(int)
df_top5.plot(kind = 'line', figsize = (14,8))
plt.title('Immigranys from  top 5 countries')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```
